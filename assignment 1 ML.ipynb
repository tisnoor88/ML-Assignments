Q1 (d) Find the most frequent value and their indice(s)
i. x = np.array([1, 2, 2, 3, 1, 1, 2])
ii. y = np.array([5, 1, 5, 3, 5, 2, 2])

```python
x = np.array([1, 2, 2, 3, 1, 1, 2])
y = np.array([5, 1, 5, 3, 5, 2, 2])
most_freq_x = np.bincount(x).argmax()
indices_x = np.where(x == most_freq_x)[0]
most_freq_y = np.bincount(y).argmax()
indices_y = np.where(y == most_freq_y)[0]
```

Q1 (f) For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
i. Sum of diagonal elements
ii. Eigen values of matrix
iii. Eigen vectors of matrix
iv. Inverse of matrix
v. Determinant of matrix

```python

import numpy as np
n_array = np.array([[55,25,15], [30,44,2],[11,45,77]])
dia = np.trace(n_array)
eig= np.linalg.eigvals(n_array)
print(dia, eig)


eig, vec= np.linalg.eig(n_array)
det= np.linalg.det(n_array)
inverse = np.linalg.inv(n_array)
print (dia,eig,vec,inverse,det)
```

Q1 (g) Multiply matrices and find covariance
i. p = [[1, 2], [2, 3]]; q = [[4, 5], [6, 7]]

```python
import numpy as np
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
product1 = np.dot(p1, q1)
cov1 = np.cov(p1.T, q1.T)

print(product1)
print(cov1)
```

ii. p = [[1, 2], [2, 3], [4, 5]]; q = [[4, 5, 1], [6, 7, 2]]

```python

import numpy as np

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
product2 = np.dot(p2, q2)
cov2 = np.cov(p2.flatten(), q2.flatten())


print(product2)
print(cov2)
```

Q1 (h) For matrices x and y, find inner, outer and cartesian product

```python
import numpy as np

x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
inner_product = np.inner(x, y)
outer_product = np.outer(x, y)
cartesian_product = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)

print("inner:", inner_product)
print("outer:" , outer_product)
print("cartesian:", cartesian_product)
```

Q2 (a) For array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Element-wise absolute value
ii. Percentiles
iii. Statistics

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_array = np.abs(array)
percentiles_flat = np.percentile(array.flatten(), [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)
mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)
mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nPercentiles (flat):", percentiles_flat)
print("Percentiles (columns):\n", percentiles_col)
print("Percentiles (rows):\n", percentiles_row)
print("\nMean (flat):", mean_flat)
print("Median (flat):", median_flat)
print("Std (flat):", std_flat)
print("\nMean (columns):", mean_col)
print("Median (columns):", median_col)
print("Std (columns):", std_col)
print("\nMean (rows):", mean_row)
print("Median (rows):", median_row)
print("Std (rows):", std_row)
```

Q2 (b) For array: a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

```python

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_vals = np.floor(a)
ceil_vals = np.ceil(a)
trunc_vals = np.trunc(a)
rounded_vals = np.round(a)
print(" Original array:", a)
print("Floor values:", floor_vals)
print("Ceil values:", ceil_vals)
print("Truncated values:", trunc_vals)
print("Rounded values:", rounded_vals)
```
